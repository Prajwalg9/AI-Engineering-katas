* Vectors

```python
arr=np.arange(1,21)
arr
```

```python
arr[9]
```

```python
arr[:5]
```

```python
arr[9:]
```

```python
arr[3::2]
```

* Matrix

```python
arr=np.arange(1,31).reshape(6,5)
arr
```

```python
arr[5] #rows
```

```python
arr[3,1] #[row,col]
```

* Slicing a specific portion from matrix:

```python
arr[0:4,1:4]
```

```python
arr[2:,3:]
```

```python
arr[:2,:3]
```

* To get specific row

```python
arr[2]
```

```python
arr[2,:] #or
```

* To get specific column

```python
arr[:,4]
```
