# Shallow Copy VS deep Copy

- Shallow Copy

```python
a = np.arange(26)
a
```

```python
b=a[5:] #shallow copy means if b changes, it won't reflect in a
b
```

```python
b[4]=333
b,a
```

- Deep Copy

```python
a = np.arange(26)
a
```

```python
b = a #Deep copy means if b changes it will reflect in a
```

```python
b[2]=1111111
a,b
```
