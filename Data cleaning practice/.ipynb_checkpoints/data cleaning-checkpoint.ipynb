```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cafe_sales.csv')
df.head(20)
```

```python
df.isnull().sum()
```

```python
df = df.replace(['UNKNOWN','ERROR'],np.nan)
df
```

```python
df=df.dropna(subset=['Item']).copy()
```

```python
df
```

```python
for col in df.columns:
    print(col,":",df[col].dtype)
```

```python
df.isnull().sum()
```

```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df.isnull().sum()
```

```python
df['Price Per Unit'] = df.groupby('Item')['Price Per Unit'].transform(
    lambda x : x.fillna(x.median())
)
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df.isnull().sum()
```

```python
for col in df.columns:
    print(col,":",df[col].dtype)
```

```python
df['Total Spent'] = df['Quantity'] * df['Price Per Unit']
df.isnull().sum()
```

```python
df
```

```python
for col in df.columns:
    print(df[col].value_counts())
    print('*'*50)
```

```python
df["Payment Method"] = df["Payment Method"].fillna("Other")
df
```

```python
df['Location'] = df['Location'].fillna("Unknown")
df
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'],errors='coerce')
```

```python
for col in df.columns:
    print(col,":",df[col].dtype)
```

```python
df['Transaction Date'] = df['Transaction Date'].ffill()
df
```
