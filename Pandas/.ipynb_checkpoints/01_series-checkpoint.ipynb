```python
import pandas as pd
import numpy as np
```

# Series

```python
#Series - One dimensional array capable of holding any type of datatype
```

```python
# from list
mylist=[19,33,55566,4.44]
pd.Series(mylist)
```

```python
# from list with indexes
idxs=['a','b','c','d']
pd.Series(mylist,index=idxs)
```

```python
# From numpy arrays
arr=np.arange(1,5)
pd.Series(arr,index=idxs)
```
