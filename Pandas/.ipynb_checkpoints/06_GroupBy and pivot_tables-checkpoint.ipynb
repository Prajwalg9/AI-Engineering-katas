```python
import numpy as np
import pandas as pd 
```

# Group By

```python
data = {
    'category':['A','B','B','A','B'],
    'stores':['s1','s1','s2','s2','s1'],
    'sales':[100,100,200,100,600],
    'quantity':[4,7,0,3,6],
    'date':['12-2024','13-2033','20-2030','10-2012','12-2020']
}
df = pd.DataFrame(data)
df
```

```python
# groupby category by sum of sales:
cat = df.groupby('category')['sales'].sum()
cat
```

```python
# groupby store by sum of sales:
store = df.groupby('stores')['sales'].sum()
store
```

```python
# groupby category,store by sum of sales:
multi = df.groupby(["category","stores"])['sales'].sum()
multi
```

# Pivot Table

```python
pd.pivot_table(df,values='sales',index='category',columns='stores') #aggFunc=mean by default it can be median,average, etc.
```
