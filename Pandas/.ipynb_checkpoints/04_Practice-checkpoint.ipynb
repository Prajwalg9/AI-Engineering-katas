# Phase 1: Subscription Pricing (Series)

# Phase 2: User Data Extraction

```python
# Here is the raw data from the engineering team:
streamify_data = {
    'UserID': [101, 102, 103, 104, 105, 106],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'Age': [24, 17, 32, 28, np.nan, 15],
    'Plan': ['Premium', 'Basic', 'Standard', 'Premium', 'Basic', 'Standard'],
    'WatchHours': [45.5, np.nan, 22.0, 50.2, 12.5, np.nan],
    'Rating': [5.0, 3.5, np.nan, 4.8, 2.0, 4.0]
}
df = pd.DataFrame(streamify_data)
df

# 2. Convert this dictionary into a Pandas DataFrame called 'df' and display it.
```

# Phase 3: Slicing and Dicing the Data

```python
# 3. The marketing team only wants to see the names and watch hours. 
# Select ONLY the 'Name' and 'WatchHours' columns.
prefered_cols = df[["Name",'WatchHours']]
prefered_cols
```

```python
# 4. We need to target adult users. 
# Filter the DataFrame to show only users whose 'Age' is greater than 18.
df[df["Age"]>18]
```

```python
# 5. EXAM CHALLENGE (Multiple Conditions): 
# Find the highly engaged premium users. 
# Filter the DataFrame for users who are on the 'Premium' plan AND watched more than 40 hours.
df[(df['Plan']=='Premium') & (df['WatchHours']>40)]
```

# Phase 4: Column Operations

```python
# 6. The engineering team says 'UserID' is no longer needed for our analysis.
# Drop the 'UserID' column PERMANENTLY from the DataFrame.
df.drop('UserID',axis=1,inplace=True)
df
```

```python
# 7. We are running a new campaign. Add a new column called 'EligibleForPromo'.
# Assign the value "Yes" to all rows in this new column for now.
df['EligibleForPromo']='yes'
df
```

# Phase 5: The Janitor Work

```python
# 8. Let's inspect the damage. 
# Write the code to find the TOTAL number of missing (NaN) values in EACH column.
df.isna().sum()
```

```python
# 9. We cannot do math with NaNs! 
# Fill the missing values in the 'WatchHours' column with the MEAN (average) of that column.
# Overwrite the column so the changes are saved.
df['WatchHours'] = df['WatchHours'].fillna(df['WatchHours'].mean())
df
```

```python
# 10. For the missing 'Age', we cannot guess their age. 
# Drop any ROW that still has a missing value in the 'Age' column.
# Make sure this change is permanent!
df['Age'].dropna(inplace=True)
```
