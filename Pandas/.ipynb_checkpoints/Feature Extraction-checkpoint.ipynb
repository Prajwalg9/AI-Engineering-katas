```python
#loading the data
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('anime.csv')
df.head()
```

# make a new column for episode count

```python
for i in [df.loc[x]["Title"] for x in range(5)]:
    print(i)
```

```python
import re
pattern = r'\((\d+)'
df["Episodes"]= df['Title'].apply(lambda x: re.findall(pattern,str(x))[0])
df.head()
```

# make a new column for time stamp

```python
for i in [df.loc[x]["Title"] for x in range(5)]:
    print(i)
```

```python
dates = df['Title'].str.extract(
    r'([A-Za-z]{3}\s\d{4})\s*-\s*([A-Za-z]{3}\s\d{4})'
)

df['timestamp'] = (
    (pd.to_datetime(dates[1], format='%b %Y').dt.year - 
     pd.to_datetime(dates[0], format='%b %Y').dt.year) * 12
    +
    (pd.to_datetime(dates[1], format='%b %Y').dt.month - 
     pd.to_datetime(dates[0], format='%b %Y').dt.month)
)
df.head()
```

# which anime has the highest score

```python
df[df['Score'] == df['Score'].max()][["Title","Score"]]
```

# give me top 5 highest scoring anime


```python
df.nlargest(5,'Score')[['Title','Score']]
```

# which anime has the highest episode count

```python
df[df['Episodes']==df['Episodes'].max()][["Title",'Episodes']]
```

# animes with top 5 episode count

```python
df.nlargest(5,"Episodes")[["Title",'Episodes']]
```

# which is the longest running anime

```python
df.nlargest(1,"timestamp")[["Title",'Episodes']]
```
