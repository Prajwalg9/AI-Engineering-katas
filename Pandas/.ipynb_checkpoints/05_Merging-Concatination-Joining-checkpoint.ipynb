```python
import numpy as np 
import pandas as pd
```

# Merging

```python
emps = {
    "id" : [1,2,3,4,5],
    "name" : ['radha','krushna','money','shubhra','varsha'],
    "dept" : ["IT",'Computers','Cloud','Data Science','AI/ML']
}
salaries = {
    "id" : [1,2,3,4,6,7] ,
    "salary" : [10000,20000,30000,40000,50000,600000] ,
    "bonous" : [5000,6000,7000,8000,900,200] ,
}
df1 = pd.DataFrame(emps)
df2 = pd.DataFrame(salaries)
```

```python
pd.merge(df1,df2,on='id',how='inner') #default is inner which takes only rows that dont have NaN
```

```python
pd.merge(df1,df2,how='outer') # outer takes all rows from both tables and fils with NaN
```

```python
pd.merge(df1,df2,how='left') #takes rows only from first table
```

```python
pd.merge(df1,df2,how='right') #takes rows only from second table
```

# Concatination

```python
table1 = {
    'A': ['A1','A2','A3'],
    'B': ['B1','B2','B3'],
    'C': ['C1','C2','C3'],
}
table2 = {
    'A': ['A4','A5','A6'],
    'B': ['B4','B5','B6'],
    'C': ['C4','C5','C6'],
}
df1 = pd.DataFrame(table1)
df2 = pd.DataFrame(table2)
df1,df2
```

```python
pd.concat([df1,df2]) #default axis is 0
```

```python
pd.concat([df1,df2],axis=1) 
```

# Joining


```python
tb1 = {
    "Name" : ['Alice','Bob','Rob']
}
tb2 = {
    "Scores" : [95,99,89]
}
df1 = pd.DataFrame(tb1,index=[1,2,3])
df2 = pd.DataFrame(tb2,index=[2,3,4])
df1,df2
```

```python
df1.join(df2) # Default how is 'inner'
```

```python
df1.join(df2,how='outer') 
```

```python
df1.join(df2,how='left')
```

```python
df1.join(df2,how='right')
```
