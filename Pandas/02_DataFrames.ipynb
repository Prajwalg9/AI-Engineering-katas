# Selection and Indexing:

```python
data= {
    'Name' : ['Jhon',"Mon","corn","Doreamon"],
    "Age" : [23,23,11,22],
    "Salary" : [2200.0,22233.0,223.300,230000.0],
    "Phone" : [7273142637,[3333333333,4444444444],3234456532,3333222244],
    "City" : ["Nashik","Mumbai","Pune","Japan"],
}
df=pd.DataFrame(data)
df
```

```python
df["Name"] # Selecting perticular column
```

```python
df[["Name","Age"]] #Selecting multiple columns
```
