## Aggregate functions

```python
data = {
    'category':['A','B','B','A','B'],
    'stores':['s1','s1','s2','s2','s1'],
    'sales':[100,100,200,100,600],
    'quantity':[4,7,0,3,6],
    'date':['12-2024','13-2033','20-2030','10-2012','12-2020']
}
df = pd.DataFrame(data)
df
```

```python
# we have aggreate functions sum, mean, median, min, max, count, std, var, prod
```

```python
df['sales'].mean()
```

```python
df['sales'].prod()
```

```python
df['sales'].count()
```

```python
# etc
```

```python
# We can also do it for multiple results
```

```python
df['sales'].agg(['mean','median', 'min', 'max', 'count', 'std', 'var','sum']) # prod can change format of values so excluded that
```

# Crostabs

```python
pd.crosstab(df['stores'],df['category'])
```

```python
pd.crosstab(df['stores'],df['quantity'])
```

```python
pd.crosstab(df['stores'],df['sales'])
```
