```python
import numpy as np 
import pandas as pd
```

# Missing Data

```python
data = {
    'A': [1,23,344,np.nan,np.nan],
    "B": [np.nan ,22,33,44,34.3],
    "C": [22,33,22,4.22,2],
    "D": [np.nan,np.nan,5,np.nan,np.nan],
    "E": [22,33,22,4.22,2],
}
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
df['A'].isna().any()
```

# Removing Null Values:

```python
df.dropna() # just a snapshot real data not changes till you use inplace=True
```

```python
df
```

# Filling missing values

```python
# Filling with zeros
```

```python
df.fillna(0)
```

```python
# Filling by column wise values
```

```python
values = {
    "A":100,
    "B":200,
    "C":300,
    "D":400,
    "E":500,
}
df.fillna(value=values)
```

```python
# Filling mean 
```

```python
df.fillna(df.mean())
```
