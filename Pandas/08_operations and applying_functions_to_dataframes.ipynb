# applynig functions to dataframe

```python
data = {
    "A":[11,223,12,55,22],
    "B":[33,444,55,22,55],
    "C":[22,11,33,44,22],
}
df = pd.DataFrame(data)
df
```

```python
def square(x):
    return x*x
df['D'] = df['A'].apply(square)
df
```

```python
# this can also be done by :
df['E'] = df['A'].apply(lambda x: x*x)
df
```
